3. Preprocessing

```python
# define columns to clean
columns_to_clean = ['_CASTHM1', '_RACE', '_INCOMG1', '_SMOKER3', '_RFBING6', '_RFHLTH', '_RFHYPE6', '_RFCHOL3']

# replace placeholder values (who answered missing = 9) with NaN and drop them
df_cleaned_filtered = df_subset.copy()
df_cleaned_filtered[columns_to_clean] = df_cleaned_filtered[columns_to_clean].replace(9, pd.NA)
df_cleaned_filtered = df_cleaned_filtered.dropna()

# check the number of remaining rows
print(f"remain rows: {df_cleaned_filtered.shape[0]}")
```

```python
# check null value in subset data
print(df_cleaned_filtered.isnull().sum())

# summary statistics
print(df_cleaned_filtered.describe())
```

After cleaning dataset, there are 273829 rows left. I decided to just go through the cleaned dataset.

```python
# encode categorical variables into numeric values for machine learning
df_encoded = pd.get_dummies(df_cleaned_filtered, 
                            columns = ['_RACE', '_SMOKER3','_RFHLTH'], drop_first = True)
binary_columns = ['_SEX', '_TOTINDA', '_RFBING6', '_RFHYPE6', '_RFCHOL3']
df_encoded[binary_columns] = df_encoded[binary_columns].apply(lambda x: x.astype('int'))

print(df_encoded.head())
```

```python
# normalize continuous variables
from sklearn.preprocessing import MinMaxScaler

continuous_columns = ['_BMI5', '_AGEG5YR', '_INCOMG1']
scaler = MinMaxScaler()
df_encoded[continuous_columns] = scaler.fit_transform(df_encoded[continuous_columns])

print(df_encoded[continuous_columns].head())
```

```python
# split the data for machine learning
from sklearn.model_selection import train_test_split

X = df_encoded.drop(columns = ['_CASTHM1'])
y = df_encoded['_CASTHM1']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, 
                                                    random_state = 819)

print(f"training set: {X_train.shape}, testing set: {X_test.shape}")
```

4. Feature selection to figure out the meaningful variables

```python
# check correlation again to check if it becomes 
# better than it from EDA after normalization
# so, change the y_train to numeric first for correlation matrix
y_train = y_train.astype(float)
correlation_matrix = X_train.corrwith(y_train)

print(correlation_matrix.sort_values(ascending=False))
```

```python
# run RFE to select significant features for machine learning
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# initialize logistic regression model
model = LogisticRegression(max_iter=1000)

# perform RFE to select top features
rfe = RFE(model, n_features_to_select=10) 
rfe.fit(X_train, y_train)

# identify selected features
selected_features = X_train.columns[rfe.support_]
print("Top features from RFE:", selected_features)

# feature ranking
ranking = pd.DataFrame({'Feature': X_train.columns, 'Rank': rfe.ranking_})
print(ranking.sort_values(by='Rank'))
```

5. Develop model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

# no balance
y_test = y_test.astype(float)

# initialize and train logistic regression model
logistic_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_model.fit(X_train[selected_features], y_train)

# make predictions
y_pred_logistic = logistic_model.predict(X_test[selected_features])

# evaluate the model
print("Logistic Regression table:")
print(classification_report(y_test, y_pred_logistic))

# calculate and display AUC score
auc_logistic = roc_auc_score(y_test, logistic_model.predict_proba(X_test[selected_features])[:, 1])
print(f"Logistic Regression AUC score: {auc_logistic:.2f}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_auc_score

# after balancing
y_test = y_test.astype(float)

# balanced regression model
logistic_model = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced') # balancing for asthma class
logistic_model.fit(X_train[selected_features], y_train)

y_pred_logistic = logistic_model.predict(X_test[selected_features])

print("Logistic Regression table::")
print(classification_report(y_test, y_pred_logistic))

auc_logistic = roc_auc_score(y_test, logistic_model.predict_proba(X_test[selected_features])[:, 1])
print(f"Logistic Regression AUC score: {auc_logistic:.2f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, roc_auc_score

# initialize Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced') 
rf_model.fit(X_train[selected_features], y_train)

# predictions
y_pred_rf = rf_model.predict(X_test[selected_features])

# evaluation table
print("Random Forest table:")
print(classification_report(y_test, y_pred_rf))

# AUC score
auc_rf = roc_auc_score(y_test, rf_model.predict_proba(X_test[selected_features])[:, 1])
print(f"Random Forest AUC score: {auc_rf:.2f}")
```

```python
from sklearn.tree import DecisionTreeClassifier

# decision Tree
dt_model = DecisionTreeClassifier(random_state=42, class_weight='balanced') 
dt_model.fit(X_train[selected_features], y_train)

y_pred_dt = dt_model.predict(X_test[selected_features])

print("Decision Tree table:")
print(classification_report(y_test, y_pred_dt))

# AUC score
auc_dt = roc_auc_score(y_test, dt_model.predict_proba(X_test[selected_features])[:, 1])
print(f"Decision Tree AUC score: {auc_dt:.2f}")
```
